```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df["event_time"], df["aqi_daily"])
plt.title("Daily AQI Over Time")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.show()
```

```python
plt.hist(df["aqi_daily"], bins=20)
plt.title("Distribution of Daily AQI")
plt.show()
```

```python
df[["pm2_5_mean","pm10_mean","no2_mean","so2_mean","co_mean"]].hist(figsize=(12,8))
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# keep only numeric columns
num_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(10,8))
sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix (Numeric Only)")
plt.show()
```

```python
df2 = df.copy()
df2["weekday"] = df2["weekday"].astype("category").cat.codes  # Mon..Sun -> 0..6 (order depends on categories)

num_df2 = df2.select_dtypes(include=[np.number])

plt.figure(figsize=(10,8))
sns.heatmap(num_df2.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix (Weekday Encoded)")
plt.show()
```

```python
plt.scatter(df["pm2_5_mean"], df["aqi_daily"])
plt.xlabel("PM2.5")
plt.ylabel("AQI")
plt.show()
```

```python
df.groupby("weekday")["aqi_daily"].mean().plot(kind="bar")
plt.title("Average AQI by Weekday")
plt.show()
```

```python
df["aqi_7day_avg"] = df["aqi_daily"].rolling(7).mean()

plt.plot(df["event_time"], df["aqi_daily"], alpha=0.5)
plt.plot(df["event_time"], df["aqi_7day_avg"], color="red")
plt.title("AQI with 7-Day Rolling Average")
plt.show()
```

```python
sns.boxplot(x=df["aqi_daily"])
plt.show()
```
